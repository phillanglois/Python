```python
#Concevoir et coder un module points qui permet de définir des points, de les comparer, d’afficher leurs coordonnées  à l’écran.
import matplotlib.pyplot as plt
from matplotlib.pyplot import*
import math 
import numpy as np
def point(x1: list[float], y1: list[float], x2: list[float],y2: list[float], n: int)-> float:
    n=len(x1)
    for i in range(n):
        for j in range(n):
            dist=math.sqrt(((x1[i]-x2[j])*(x1[i]-x2[j]))+((y1[i]-y2[j])*(y1[i]-y2[j])))
            print(dist)
            x1.append(x1[i])
            y1.append(y1[i])
            x2.append(x2[j])
            y2.append(y2[j])
            listpoints=np.array([x1,y1,x2,y2])
    figure("Nuage de points (version 1)")
    # Titre du graphique
    title('Nuage de points')
    # Légende des deux axes
    xlabel('axe des X')                     
    ylabel('axe des Y')
    # Création d'une image contenant la figure
    savefig('NuagePoint1.jpg')
    grid()           # Affiche une grille
    plt.legend()         # Affiche la légende
    #show()           # Affiche la figure
    plt.plot(x1, y1,'o')
    plt.plot(x2, y2,'o')
    plt.show() 
    #print(listpoints)
    return listpoints
point([1,0],[0,1],[2,0],[1,2],4)
```

```python
#Compléter ce module de façon à pouvoir enregistrer un nuage de points de taille arbitraire dans un fichier de texte. La taille du nuage sera aussi enregistrée dans ce fichier.
#Les fichiers data_nuage_10.txt et data_nuage_100.txt fournis dans l’archive 3-data-nuage.zip sont des exemples de tels fichiers.
from matplotlib.pyplot import*
import random
import pandas as pd
def write_nuage(n: int)-> list[float]:
    X = [0 for i in range(n)]
    Y = [0 for i in range(n)]
    for i in range(n):
        X[i] = random.uniform(0,1)
        Y[i] = random.uniform(0,1)
        X.append(X[i])
        Y.append(Y[i])
    #X.sort()
    #Y.sort()
    figure("Nuage de points Trié")
    # Titre du graphique
    title('Nuage de points en [0,1]x[0,1] Trié')
    # Légende des deux axes
    xlabel('axe des X')                     
    ylabel('axe des Y')
    plt.plot(X,Y, 'o')
    # Création d'une image contenant la figure
    savefig('NuagePoint2.jpg')
    grid()           # Affiche une grille
    legend()         # Affiche la légende
    show()           # Affiche la figure
    x=np.transpose([X])
    y=np.transpose([Y])
    #print("x",x)
    #print("y",y)
    xdf=pd.DataFrame(x,columns=['x'])
    ydf=pd.DataFrame(y,columns=['y'])
    #print("xdf",xdf)
    #print("ydf",ydf)
    result=pd.concat([xdf,ydf],axis=1)
    f = open ("100points.txt", "w")
    f.write (str(n))
    f.write ("\n") 
    f.write (str(result))                          
    f.close () 
    return x,y
write_nuage(100)
```

```python
from matplotlib.pyplot import*

def orientation (p ,q ,r ):
    xpq , ypq = q[0] -p[ 0], q[1 ]-p [1]
    xpr , ypr = r[0] -p[ 0], r[1 ]-p [1]
    orient = xpq * ypr - xpr * ypq
    if orient > 0:
        return 1
    elif orient < 0:
        return -1
    else :
        return 0
orientation([0, 0], [2, 1], [-1, 2])
```

```python
def convex(L):
    n = len(L)
    EnvConvexe = []
    for i in range(n):
        for j in range(n):
            Listeorientation = []
            if i != j:
                for k in range(n):
                    if k != i and k != j:
                        Listeorientation.append(orientation(L[i], L[j], L[k]))
                a = Listeorientation[0]
                sommet = True
                for v in Listeorientation:
                    if v != a:
                        sommet = False
                if sommet and (i not in EnvConvexe):
                    EnvConvexe.append(i)
                if sommet and (j not in EnvConvexe):
                    EnvConvexe.append(j)
    return EnvConvexe
L = [[0, 0], [2, 1], [-1, 2]]
Enveloppe = convex (L )
print(Enveloppe)# PhL

plt.plot([L[i][0] for i in Enveloppe], [L[i][1] for i in Enveloppe ])
plt.show()
```
